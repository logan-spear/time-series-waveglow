```python
import waveglow_model as model
import pandas as pd
import numpy as np
```

```python
flows_list = [4, 6, 8, 15]
layers_dilation_pairs = [
    [2, [1,1]], 
    [2, [1,2]], 
    [4, [1,1,1,1]], 
    [4, [1,1,2,2]]
]
n_group_list = [96, 48, 24]

early_every_early_size_pairs = [
    [99, 99],
    [1, 8], 
#     [1, 12],
    [2, 12],
    [2, 24]
]

kernel_list = [3,5]

config_list = []

for n_flows in flows_list:
    for layers_dilation in layers_dilation_pairs:
        for n_group in n_group_list:
            for early_early in early_every_early_size_pairs:
                for kern in kernel_list:
                    config = {'n_flows': n_flows, 
                              'n_group': n_group, 
                              'n_early_every': early_early[0], 
                              'n_early_size': early_early[1], 
                              'n_layers': layers_dilation[0],  
                              'dilation_list': layers_dilation[1], 
                              'kernel_size': kern, 
                              'learning_rate': 0.001, 
                    }
                    
                    # Try instantiating model to find ones with invalid combinations of params
                    try:
                        print("********************************")
                        net = model.WaveGlow(
                            n_context_channels=96, 
                            n_flows=config['n_flows'], 
                            n_group=config['n_group'], 
                            n_early_every=config['n_early_every'],
                            n_early_size=config['n_early_size'],
                            n_layers=config['n_layers'],
                            dilation_list=config['dilation_list'],
                            n_channels=96,
                            kernel_size=config['kernel_size'], 
                            use_cuda=False
                        )
                        
                        config_list.append(config)
                    except:
                        print('--------------------------------------')
                        print("Skipped for config: ", config)
                        print('--------------------------------------')
                        
                    



print(len(config_list))
print(config_list[0])

import random
random.shuffle(config_list)
print(config_list[0])

import pickle

with open("list_of_232_configs", "wb") as fp:
    pickle.dump(config_list, fp)

with open("list_of_232_configs", "rb") as fp:
    test_list = pickle.load(fp)

test_list == config_list
```
